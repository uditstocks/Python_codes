# 4. Count no of vovles

```python
vovle = ["a" ,"e" ,"i" , "o" ,"u"]
word = str(input("Enter word you want to count the vovle: "))

count = 0
for character in word:
    if character in vovle:
        count += 1

print(count)
```

# 5. Count no of consonent 

```python
vovle = ['a' , 'e', 'i' , 'o', 'u']
word = str(input("Enter a word: "))

count = 0
for character in word:
    if not character in vovle:
        count += 1

print(count)
```

# 11. Comparing two strings for ANAGRAMS

Anagrams are words or strings that contain the same characters with the same frequency, just arranged in a different order.

```python
str1 = "Listen"
str2 = "Silent"
# Normalize → Sort → Compare = Anagram check 

str1 = str1.replace(" ", "").upper()
str2 = str2.replace(" ", "").upper()

if sorted(str1) == sorted(str2):
    print("True")
else:
    print("False")
```

# 13. Counting Digits, Letters and spaces in a string

```python
# importing regular exprersion library
import re

name = "Python is 1"

digitcount = re.sub(r"[^0-9]", "",name)
lettercount = re.sub(r"[^a-zA-Z]", "", name)
spacecount = re.findall(r"[ \s]", name)

print(len(digitcount))
print(len(lettercount))
print(len(spacecount))
```

# 14. Create a generator to produce first n prime numbers

```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, num):
        if num%i == 0:
            return False
    return True

def prime_generator(n):
    num = 2
    while n:
        if is_prime(num):
            yield num
            n-=1
        num+=1
        
```

```python
x = int(input("enter how many prime numbers you want: "))
result = prime_generator(x)
for e in result:
    print(e, end = " ")
```

```python
def prime_no(n):
    Flag = False
    if n < 2:
        return Flag
    else:
        for i in range(2,n):
            if n%i == 0:
                return Flag
        Flag = True
        return Flag
num = int(input("enter a number: "))
prime_no(num)
```

# 18. Monkey patching

Monkey patching is replace a function objects with a new function object. so the function is now pointing to new function object.
mostly used when you want to replace a function for testing purpose

```python
class Test:
    def __init__(self, x):
        self.a = x
    def get_data(self):
        print("send code to fetch data from database")
    def f1(self):
        self.get_data()
    def f2(self):
        self.get_data()

t1 = Test(4)
print("before monkey patching")
t1.f1()
t1.f2()

# for testing purpose we make new fun
def get_new_data(self):
    print("fetch data from test data")
Test.get_data = get_new_data
print("After monkey patching")
t1.f1()
t1.f2()
```

# 19. Extract int type values from a list of heterogeneous elements

```python
l1 = ["abcd", 44.33, 33, 4+6+9j, "b", 55, 65.3, "90", 190]
print(l1)

l2 = []
for e in l1:
    if type(e) == int:
        l2.append(e)
print(l2)
```

# 20. Default arguments in function
