# 14. Create a generator to produce first n prime numbers

```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, num):
        if num%i == 0:
            return False
    return True

def prime_generator(n):
    num = 2
    while n:
        if is_prime(num):
            yield num
            n-=1
        num+=1
        
```

```python
x = int(input("enter how many prime numbers you want: "))
result = prime_generator(x)
for e in result:
    print(e, end = " ")
```

```python
def sand(n):
    for i in range(n):
        # spaces (increase)
        for j in range(i):
            print(" ",end="")
        # stars (decrease)
        for j in range(n-i):
            print("*",end=" ")
        print()

    for i in range(n):
        for j in range(i,n-1):
            print(" ", end="")
        for j in range(i+1):
            print("*", end=" ")
        print()
```

# 46. Convert a given Roman numeral string into its equivalent integer value following addition and subtraction rules.

```python
class Solution(object):
    def romanToInt(self, s):
        values = {
            'I': 1, 'V': 5, 'X': 10,
            'L': 50, 'C': 100,
            'D': 500, 'M': 1000
        }

        total = 0

        for i in range(len(s)):
            if i < len(s)-1 and values[s[i]] < values[s[i+1]]:
                total -= values[s[i]]
            else:
                total += values[s[i]]

        return total
        return total
```

# 48. Binary Search

1. Works only on a sorted array
2. Divide & Conquer:
    Check middle element → eliminate half of the search space each step.
3. Pointer Movement:
    If target < mid → search left half
    If target > mid → search right half

```python
def binarysearch(arr, element):
    start = 0
    end = len(arr) - 1
    while(start <= end):
        mid = (start + end) // 2
        if arr[mid] == element:
            return mid
        elif arr[mid] < element:
            start = mid + 1
        else:
            end = mid - 1
```

```python
arr = [1,2,3,4,5]
index = binarysearch(arr, 5)
print(index)
```

# 49. Selection Sort

1. For each position, find the smallest element from the rest and place it there.
2. Works on unsorted array
3. Goal: Put the correct element at position i in every step
4. Strategy: Find minimum from remaining array → swap it to front

```python
def selectionshort(arr):
    length = len(arr)
    #put the correct element at "i"th position 
    for i in range(length - 1):
        minIndex = i
        #calculating the index of minimum element
        for j in range(i + 1, length):
            if(arr[j] < arr[minIndex]):
                minIndex = j
        arr[i],arr[minIndex] = arr[minIndex],arr[i]
    
arr = [1,3,6,2,4,9,8,5,7,45]
selectionshort(arr)
print(arr)
```

# 50. Bubble Sort 

1. Works on any array
2. Repeatedly compares adjacent elements
3. If left > right → swap
4. After each pass → largest element moves to the end

```python
def bubbleSort(arr):
    length = len(arr)
    #i is for n-1 rounds
    #j is for in each iteration u need to go till length - 1- i position
    for i in range(length - 1):
        for j in range(length - 1 - i):
            if (arr[j] > arr[j + 1]):
                arr[j],arr[j + 1] = arr[j + 1], arr[j]
                
        

arr = [1,11,8,5,20,2]
bubbleSort(arr)
print(arr)
```

# 51. Insertion Sort

1. Keep shifting bigger elements right until you find the correct place
2. Works on any array
3. Builds a sorted portion step-by-step
4. Take one element → insert it at correct position

```python
def insertionSort(arr):
    length = len(arr)
    for i in range(1, length):
        j = i - 1
        temp = arr[i]
        #Shifting elements till this condition holdes
        while(j>=0 and arr[j] > temp):
            arr[j + 1] = arr[j]
            j = j - 1
        #J +1 is correct position for ith element
        arr[j + 1] = temp

arr = [1,11,8,5,20,2,12]
insertionSort(arr)
print(arr)
```

# 52. Merge two Sorted Arrays

1. Two pointers → always pick the smaller front element
2. No sorting needed → arrays are already sorted, just merge smartly
3. Linear time → O(n + m), super efficient

```python
def mergeSortedArrays(arr1, arr2):
    i = 0
    j = 0
    len1 = len(arr1)
    len2 = len(arr2)
    arr = []

    while ((i < len1) and (j < len2)):
        if(arr1[i] < arr2[j]):
            arr.append(arr1[i])
            i = i + 1
        else:
            arr.append(arr2[j])
            j = j + 1
    while (i < len1):
        arr.append(arr1[i])
        i = i + 1
    while (j < len2):
        arr.append(arr2[j])

        j = j + 1
    return arr
        
arr1 =[1, 3, 5, 8, 12, 19, 21]
arr2 =[2, 4, 6, 9, 11, 17, 22]
arr = mergeSortedArrays(arr1,arr2)
print(arr)
```

# 53. Binary search using recursion

```python
def binarySearch(a, x, si, ei): #here "a" is the list and "x" is the item to find
    if si > ei:
        return -1

    mid = (si + ei)//2
    if a[mid] == x:
        return mid
    elif a[mid] > x:
        return binarySearch(a, x, si, mid - 1)
    else:
        return binarySearch(a, x, mid + 1, ei)

y = binarySearch([1,2,3,4,5,6,7,8,9], 3, 0, 8)
print(y)
```

# 55. Abstract Classes & Abstraction (ABC)

### What's happening here?

- `Shape` is an **abstract class** - it cannot be instantiated directly.
- `@abstractmethod` forces every subclass to implement `area()` and `perimeter()`. If they don't, Python raises a `TypeError`.
- ABC can also have **concrete methods** (like `describe()`) unlike a pure interface.
- This enforces a **contract**: any class claiming to be a `Shape` must define how it calculates area and perimeter.

> **Interview angle:** "Abstraction hides *how* something works and exposes *what* it does. ABC enforces that contract at the class level."

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

    def describe(self):  # concrete method — shared by all subclasses
        print(f"I am a shape with area = {self.area():.2f}")


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14159 * self.radius ** 2

    def perimeter(self):
        return 2 * 3.14159 * self.radius


class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)


c = Circle(5)
r = Rectangle(4, 6)

print(f"Circle    -> Area: {c.area():.2f}, Perimeter: {c.perimeter():.2f}")
c.describe()

print(f"Rectangle -> Area: {r.area()}, Perimeter: {r.perimeter()}")

# Uncommenting below raises TypeError — abstract class can't be instantiated
# s = Shape()
```

# 56. MRO - Method Resolution Order

### What's happening here?

- Python resolves method lookup using **C3 Linearization** - left to right, depth-first, no class visited twice.
- `D(B, C)` → MRO is `[D, B, C, A, object]`. So `d.greet()` calls `B.greet()` (B comes before C).
- `super()` doesn't just call the parent — it calls the **next class in the MRO chain**, which is why it's safe in multiple inheritance.
- Always use `ClassName.__mro__` to debug which method will get called.

> **Interview angle:** "Diamond problem in Python is solved by MRO. Python guarantees each class appears only once in the chain, and `super()` respects that order."

```python
class A:
    def greet(self):
        print("Hello from A")

class B(A):
    def greet(self):
        print("Hello from B")

class C(A):
    def greet(self):
        print("Hello from C")

class D(B, C):  # Diamond: D -> B -> C -> A -> object
    pass


d = D()
d.greet()   # Which greet() gets called?

print(f"MRO: {[cls.__name__ for cls in D.__mro__]}")


# Bonus: super() walks the MRO chain
class E(B, C):
    def greet(self):
        super().greet()   # calls B.greet() → next in MRO
        print("Hello from E")

print("---")
e = E()
e.greet()
```

# 57. Decorators

A decorator wraps a function to add behavior without changing its code
Think of it like a gift wrapper — the gift (function) stays the same inside, but the wrapper adds something extra on the outside. In Python, a decorator is itself a function that takes a function as input and returns a new (wrapped) function.

## Core Examples

### 1. Timer - measure execution time

```python
# measure execution time
import time
from functools import wraps

def timer(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f"{func.__name__} took {(end-start):.4f} seconds")
        return result
    return wrapper

@timer
def slow_task():
    time.sleep(1)
    return "Done"

slow_task()
```

### 2. Validator - check inputs before running|

```python
# check inputs before running
def validate_positive(func):
    @wraps(func)
    def wrapper(n):
        if n < 0:
            raise ValueError(f"Expected positive number, got {n}")
        return func(n)
    return wrapper

@validate_positive
def square_root(n):
    return n ** 0.5

print(square_root(16))    # 4.0
print(square_root(-4))    # raises ValueError
```

### 3. Logger - track when functions are called

```python
from functools import wraps

def logger(func):
    @wraps(func)               # preserves original function name & docstring
    def wrapper(*args, **kwargs):
        print(f"Calling: {func.__name__} | args={args}")
        result = func(*args, **kwargs)
        print(f"Returned: {result}")
        return result
    return wrapper

@logger
def add(a, b):
    return a + b

add(3, 5)
```

### 1. Authentication check (used in Flask/Django)

```python
def login_required(func):
    @wraps(func)
    def wrapper(user, *args, **kwargs):
        if not user.get("logged_in"):
            print("Access denied! Please log in.")
            return None
        return func(user, *args, **kwargs)
    return wrapper

@login_required
def dashboard(user):
    print(f"Welcome, {user['name']}!")

user1 = {"name": "Rahul", "logged_in": True}
user2 = {"name": "Guest", "logged_in": False}

dashboard(user1)    # Welcome, Rahul!
dashboard(user2)    # Access denied!
```

### 2. Retry on failure (network/API calls)

```python
import time
from functools import wraps

def retry(times=3, delay=1):         # decorator with arguments
    def decorator(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            for attempt in range(1, times+1):
                try:
                    return func(*args, **kwargs)
                except Exception as e:
                    print(f"Attempt {attempt} failed: {e}")
                    time.sleep(delay)
            print("All attempts failed.")
        return wrapper
    return decorator

@retry(times=3, delay=0.5)
def fetch_data():
    raise ConnectionError("Server down")

fetch_data()
```

### 3. Cache / memoization (speed up repeated calls)

```python
from functools import lru_cache

@lru_cache(maxsize=128)     # built-in caching decorator
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

print(fibonacci(10))   # 55
print(fibonacci(50))   # V
```

### Chaining decorators

When you stack decorators, they are applied bottom-up - the one closest to the function runs first. The order matters.

```python
from functools import wraps

def bold(func):
    @wraps(func)
    def wrapper():
        return "" + func() + ""
    return wrapper

def italic(func):
    @wraps(func)
    def wrapper():
        return "" + func() + ""
    return wrapper

@bold          # applied second (outer)
@italic        # applied first (inner)
def greet():
    return "Hello"

print(greet())
# Equivalent to: bold(italic(greet))()
```

### Class-based decorator
Decorators can also be implemented as classes using __init__ and __call__.

```python
class CountCalls:
    def __init__(self, func):
        self.func = func
        self.count = 0

    def __call__(self, *args, **kwargs):
        self.count += 1
        print(f"Call #{self.count} to {self.func.__name__}")
        return self.func(*args, **kwargs)

@CountCalls
def say_hi():
    print("Hi!")

say_hi()
say_hi()
say_hi()
```

### These are the most commonly asked decorator questions in Python interviews at all levels.

Q1. What is a decorator in Python?

Answer: A decorator is a design pattern in Python that allows you to add functionality to an existing
function without modifying its code. 
It is a function that takes another function as an argument, wraps it inside an inner function, and returns the wrapper.
The @syntax is syntactic sugar for reassigning the function to the decorated version.

Q2. Why do we use @wraps when writing a decorator?

Answer: When you wrap a function, Python replaces the original function object with the wrapper.
This means the original function's __name__, __doc__, and other metadata are lost.
Using @wraps(func) from the functools module copies those attributes from the original function to the wrapper - which is important for debugging and documentation.

```python
def dec(func):
    @wraps(func)              # without this: wrapper.__name__ = "wrapper"
    def wrapper(*args, **kwargs):   # with this: wrapper.__name__ = "add"
        return func(*args, **kwargs)
    return wrapper
```

Q3. Can a function have multiple decorators? What is the execution order?

Answer: Yes. Multiple decorators are applied bottom-up (innermost first).
But they execute top-down at call time.
So @bold @italic on a function means: italic is applied first during decoration, but bold's wrapper runs first when the function is called.

Q4. What is the difference between @staticmethod and @classmethod?

Answer:
@staticmethod — no access to class (cls) or instance (self). It's just a regular function scoped to the class namespace.

@classmethod — receives the class (cls) as the first argument instead of self. Used to create factory methods or access class-level data.

```python
class MyClass:
    count = 0

    @classmethod
    def get_count(cls):        # accesses class
        return cls.count

    @staticmethod
    def utility(x):            # no class/instance access
        return x * 2
```

Q6. What is @property decorator?

Answer: @property turns a method into a readable attribute.
It allows you to add validation or computed logic behind an attribute-style access.
You pair it with @name.setter to allow setting the value.

```python
class Circle:
    def __init__(self, radius):
        self._radius = radius

    @property
    def area(self):
        return 3.14 * self._radius ** 2

c = Circle(5)
print(c.area)   # 78.5 — called like attribute, not method
```

### The 4 pillars of OOP - memorise these for every interview

```python
1. — Encapsulation
Hiding internal data; exposing only what's needed
Like a capsule pill — you take it without knowing what's inside.
In Python, prefix attributes with _ (protected) or __ (private) to restrict access.
```

```python
2. — Inheritance
Child class reuses and extends a parent class
Like a child inheriting traits from a parent.
The child gets all of the parent's methods and can add its own or override existing ones.
```

```python
3. — Polymorphism
Same method name, different behavior per class
Like a "speak" command — a dog barks, a cat meows, a cow moos.
Same interface, different implementation. Achieved via method overriding in Python.
```

```python
4. — Abstraction
Hiding complexity; showing only essential details
Like driving a car — you use the steering wheel and pedals without knowing how the engine works.
In Python, done via abstract base classes (abc module).
```

### ~ Inheritance - child class extends parent class

The child class gets all methods and attributes of the parent. Use super() to call the parent's __init__ without repeating code.

```python
# Single inheritance
class Animal:
    def __init__(self, name, sound):
        self.name  = name
        self.sound = sound

    def speak(self):
        return f"{self.name} says {self.sound}!"

    def breathe(self):
        return f"{self.name} breathes air."

class Dog(Animal):              # inherits from Animal
    def __init__(self, name):
        super().__init__(name, "Woof")   # call parent __init__

    def fetch(self):            # new method only in Dog
        return f"{self.name} fetches the ball!"

d = Dog("Bruno")
print(d.speak())                # inherited from Animal
print(d.breathe())              # inherited from Animal
print(d.fetch())                # Dog-specific method
print(isinstance(d, Animal))    # True — Dog IS-A Animal
```

```python
# Multi-level inheritance
class Vehicle:
    def start(self):
        return "Vehicle started"

class Car(Vehicle):
    def drive(self):
        return "Car is driving"

class ElectricCar(Car):         # inherits from Car → Vehicle
    def charge(self):
        return "Charging battery"

ec = ElectricCar()
print(ec.start())               # from Vehicle (grandparent)
print(ec.drive())               # from Car (parent)
print(ec.charge())              # own method
print(ElectricCar.__mro__)      # method resolution order
```

```python
# Multiple inheritance
class Flyable:
    def fly(self):
        return "I can fly!"

class Swimmable:
    def swim(self):
        return "I can swim!"

class Duck(Flyable, Swimmable):  # inherits from TWO classes
    def quack(self):
        return "Quack!"

d = Duck()
print(d.fly())
print(d.swim())
print(d.quack())
```

### BankAccount example

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner    = owner        # public
        self._limit   = 50000        # protected
        self.__balance = balance     # private (name-mangled)

    def deposit(self, amt):
        if amt > 0:
            self.__balance += amt

    def withdraw(self, amt):
        if amt > self.__balance:
            return "Insufficient funds"
        self.__balance -= amt
        return f"Withdrawn ₹{amt}"

    def get_balance(self):           # getter — controlled access
        return self.__balance

acc = BankAccount("Rahul", 10000)
acc.deposit(5000)
print(acc.get_balance())             # 15000 — via getter
print(acc.withdraw(3000))            # Withdrawn ₹3000
print(acc.owner)                     # public — ok
print(acc._limit)                    # works but discouraged
print(acc.__balance)                 # AttributeError!
```

### Example Pythonic getters & setters

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.__salary = salary

    @property
    def salary(self):               # getter
        return self.__salary

    @salary.setter
    def salary(self, value):        # setter with validation
        if value < 0:
            raise ValueError("Salary can't be negative")
        self.__salary = value

e = Employee("Priya", 50000)
print(e.salary)                     # 50000 — getter
e.salary = 60000                    # setter
print(e.salary)                     # 60000
```

### ~ Polymorphism - one interface, many forms

The same method name behaves differently depending on the object calling it.
This makes code flexible and extensible - you can add new classes without changing existing logic.
Method overriding

### Method overriding

```python
class Shape:
    def area(self):
        return 0

    def describe(self):
        return f"I am a {type(self).__name__} with area {self.area()}"

class Circle(Shape):
    def __init__(self, r):
        self.r = r
    def area(self):              # overrides parent method
        return round(3.14 * self.r ** 2, 2)

class Rectangle(Shape):
    def __init__(self, w, h):
        self.w = w; self.h = h
    def area(self):              # overrides parent method
        return self.w * self.h

class Triangle(Shape):
    def __init__(self, b, h):
        self.b = b; self.h = h
    def area(self):
        return 0.5 * self.b * self.h

shapes = [Circle(5), Rectangle(4, 6), Triangle(3, 8)]
for s in shapes:
    print(s.describe())          # same method, different result
```

### Duck typing (Python-style polymorphism)

```python
class Dog:
    def speak(self):
        return "Woof!"

class Cat:
    def speak(self):
        return "Meow!"

class Robot:
    def speak(self):
        return "Beep boop."

def make_it_speak(thing):        # no type check needed
    print(thing.speak())

for creature in [Dog(), Cat(), Robot()]:
    make_it_speak(creature)
```

### ~ Special methods that Python calls automatically

Methods wrapped in double underscores (__method__) are called "dunder" methods. 
They let your objects work with Python's built-in operations like print(), len(), +, ==, and more.

```python
# A Python class demonstrating operator overloading and built-in function customization using dunder methods.
class Book:
    def __init__(self, title, author, pages):
        self.title  = title
        self.author = author
        self.pages  = pages

    def __str__(self):           # called by print() and str()
        return f'"{self.title}" by {self.author}'

    def __repr__(self):          # called in REPL, debugging
        return f'Book("{self.title}", "{self.author}", {self.pages})'

    def __len__(self):           # called by len()
        return self.pages

    def __eq__(self, other):     # called by ==
        return self.title == other.title

    def __lt__(self, other):     # called by <  (enables sorting)
        return self.pages < other.pages

    def __add__(self, other):    # called by +
        combined = self.pages + other.pages
        return f"Combined pages: {combined}"

b1 = Book("Python 101", "Rahul", 350)
b2 = Book("OOP Guide", "Priya", 200)

print(b1)                        # uses __str__
print(len(b1))                   # uses __len__
print(b1 == b2)                  # uses __eq__
print(b2 < b1)                   # uses __lt__
print(b1 + b2)                   # uses __add__
books = [b1, b2]
print(sorted(books))             # sorts using __lt__
```

@staticmethod and @classmethod

```python
# A Python class demonstrating the difference between instance methods, class methods, and static methods.
class Temperature:
    unit = "Celsius"

    def __init__(self, value):
        self.value = value

    @classmethod
    def set_unit(cls, unit):         # accesses the class, not instance
        cls.unit = unit

    @staticmethod
    def celsius_to_fahrenheit(c):    # utility — no class or instance needed
        return (c * 9/5) + 32

    def display(self):
        return f"{self.value}° {Temperature.unit}"

t = Temperature(100)
print(t.display())                   # 100° Celsius
Temperature.set_unit("Kelvin")
print(t.display())                   # 100° Kelvin
print(Temperature.celsius_to_fahrenheit(100))  # 212.0
```

### ~ Interview Q&A

These are the most commonly asked OOP questions in Python interviews 

Q1. What are the 4 pillars of OOP? Explain each briefly.

ANS: 
1. Encapsulation - bundling data and methods together and restricting direct access to internal state (using _ / __).
2. Inheritance - a child class acquires properties and behavior of a parent class, enabling code reuse.
3. Polymorphism - the same method name behaves differently in different classes (method overriding, duck typing).
4. Abstraction - hiding implementation details and exposing only the necessary interface (via ABC and @abstractmethod).

Q2. What is the difference between __str__ and __repr__?

ANS:
__str__ is the human-readable string for end users - called by print() and str().
__repr__ is the developer/debug string - called in the REPL, logs, and when __str__ is not defined.

Rule of thumb: __repr__ should return a string that could recreate the object.

Q3. What is the difference between @classmethod and @staticmethod?

ANS: 
@classmethod receives cls (the class itself) as its first argument  used to access or modify class-level state, and to write factory methods.

@staticmethod receives no implicit first argument  it's just a utility function that logically belongs in the class namespace but doesn't need class or instance access.

Q4. What is method overriding vs method overloading?

ANS:
Method overriding - a child class defines a method with the same name as the parent, replacing the parent's behavior. Python fully supports this.

Method overloading - same method with different parameter counts. Python does NOT natively support overloading. The last definition wins. You can simulate it using default parameters or *args.
