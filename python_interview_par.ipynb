# 11. Comparing two strings for ANAGRAMS

Anagrams are words or strings that contain the same characters with the same frequency, just arranged in a different order.

```python
str1 = "Listen"
str2 = "Silent"
# Normalize → Sort → Compare = Anagram check 

str1 = str1.replace(" ", "").upper()
str2 = str2.replace(" ", "").upper()

if sorted(str1) == sorted(str2):
    print("True")
else:
    print("False")
```

# 13. Counting Digits, Letters and spaces in a string

```python
# importing regular exprersion library
import re

name = "Python is 1"

digitcount = re.sub(r"[^0-9]", "",name)
lettercount = re.sub(r"[^a-zA-Z]", "", name)
spacecount = re.findall(r"[ \s]", name)

print(len(digitcount))
print(len(lettercount))
print(len(spacecount))
```

# Create a generator to produce first n prime numbers

```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, num):
        if num%i == 0:
            return False
    return True

def prime_generator(n):
    num = 2
    while n:
        if is_prime(num):
            yield num
            n-=1
        num+=1
        
```

```python
x = int(input("enter how many prime numbers you want: "))
result = prime_generator(x)
for e in result:
    print(e, end = " ")
```

```python
def prime_no(n):
    Flag = False
    if n < 2:
        return Flag
    else:
        for i in range(2,n):
            if n%i == 0:
                return Flag
        Flag = True
        return Flag
num = int(input("enter a number: "))
prime_no(num)
```
